# UK Crime Scene Invest

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import glob
```

```python
# Read UK data

def process_MOPAC_csv_files(file_path):
    # Read the CSV file
   
    df = pd.read_csv(file_path, index_col=False)

    return df
```

```python
# Street crimes all of UK

b = True
for file_path in glob.glob('\\Users\\Rhahim\\Documents\\TU_Eindhoven\\2023\\Q4\\DC2\\CSV\\2024-02\\*street.csv' ): 
    #if point_year in range(start_year, end_year+1):
    #files_to_process = f"\\Users\\Casey\\Dash\\JupyNote\\data\\Stats19\\collision_{point_year}.csv"
    #if file_path == files_to_process: 
    if file_path != '':
        #print(file_path) 
        _df_street = process_MOPAC_csv_files(file_path)
        if b == True:
            df_street = _df_street.copy()
            b = False    
        else:
            df_street= pd.concat([df_street, _df_street], ignore_index=True)
        
df_street.to_csv('\\Users\\Rhahim\\Documents\\TU_Eindhoven\\2023\\Q4\\DC2\\kc_street.csv')
```

```python

filtered_df = df_street[df_street["Crime type"] != 'Anti-social behaviour'] # Anti-social behaviour does not have a last outcome

# Perform the aggregation similar to the SQL query
d3_agg = filtered_df.groupby(['Month', 'Last outcome category']).size().reset_index(name='count')

# Sort the DataFrame by count in descending order
d3_agg = d3_agg.sort_values(by='count', ascending=False)

# Print the resulting DataFrame
d3_agg.head()
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(d3_agg['Last outcome category'], d3_agg['count'], color='skyblue')
plt.xlabel('Last Outcome Category')
plt.ylabel('Count')
plt.title('Counts of Last Outcome Categories in February 2024')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
filtered_df = df_street[df_street["Crime type"] != 'Anti-social behaviour']

# Perform the aggregation similar to the SQL query
d4_agg = filtered_df.groupby(['Month', 'Crime type', 'Last outcome category']).size().reset_index(name='count')

# Sort the DataFrame by count in descending order
d4_agg = d4_agg.sort_values(by='count', ascending=False)
```

```python
def plot_outcome_category(data, outcome_category):
    # Filter the data based on the outcome category
    plot = data[data['Last outcome category'] == outcome_category].copy()

    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    plt.bar(plot['Crime type'], plot['count'], color='skyblue')
    plt.xlabel('Last Outcome Category')
    plt.ylabel('Count')
    plt.title(f'Crime types City of London {outcome_category} in February 2024')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
```

```python
last_outcome = 'Under investigation'
plot_outcome_category(d4_agg, last_outcome)
```

```python
last_outcome = 'Investigation complete; no suspect identified'
plot_outcome_category(d4_agg, last_outcome)
```
