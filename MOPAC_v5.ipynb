```python
filtered_df = df_street[df_street["Crime type"] != 'Anti-social behaviour']

# Perform the aggregation similar to the SQL query
d4_agg = filtered_df.groupby(['Month', 'Crime type', 'Last outcome category']).size().reset_index(name='count')

# Sort the DataFrame by count in descending order
d4_agg = d4_agg.sort_values(by='Month')
#d4_agg.to_csv('\\Users\\Casey\\Documents\\TU_Eindhoven\\2023\\Q4\\DC2\\d4_agg.csv')
```

```python
import matplotlib.pyplot as plt

def plot2_outcome_category(data, crime_type, outcome_category):
    # Filter the data based on the outcome category
    plot = data[(data['Last outcome category'] == outcome_category) & (data['Crime type'] == crime_type)].copy()

    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    #plt.bar(plot['Crime type'], plot['count'], color='skyblue')
    plt.bar(plot['Month'], plot['count'], color='skyblue')
    #plt.xlabel('Last Outcome Category')
    plt.xlabel('Years')
    #plt.ylabel('Count')
    plt.title(f'City of London {crime_type} {outcome_category} ')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
```

```python
# Define the function
def plot2_outcome_category_wrapper(crime_type, outcome_category): 
    plot2_outcome_category(d4_agg, crime_type, outcome_category)

# Define dropdown widgets for crime_type and outcome_category 
crime_type_dropdown = widgets.Dropdown(
    options=['Violence and sexual offences', 'Theft from the person', 'Other theft', 'Shoplifting', 'Public order', 'Drugs',
            'Criminal damage and arson', 'Burglary', 'Other crime', 'Vehicle crime', 'Bicycle theft', 'Possession of weapons',
            ' Robbery'], 
    description='Crime Type:',
    disabled=False,
)

outcome_category_dropdown = widgets.Dropdown(
    options=['Action to be taken by another organisation','Awaiting court outcome','Court result unavailable',
	    'Formal action is not in the public interest','Further investigation is not in the public interest',
'Investigation complete; no suspect identified','Local resolution','Offender given a caution','Offender given a drugs possession warning',
'Offender given penalty notice','Status update unavailable','Suspect charged as part of another case','Unable to prosecute suspect', 
'Under investigation'], 
    description='Outcome:',
    disabled=False,
)

# Call the function when the dropdown values change
widgets.interactive(plot2_outcome_category_wrapper, crime_type=crime_type_dropdown, outcome_category=outcome_category_dropdown)
```

```python
def plot3_gender(data, gender, od_ethnicity):
    # Filter the data based on the outcome category
    plot = data[(data['Gender'] == gender) & (data['Officer-defined ethnicity'] == od_ethnicity)].copy()

    # Plot the bar chart
    plt.figure(figsize=(10, 6))
    #plt.bar(plot['Crime type'], plot['count'], color='skyblue')
    plt.bar(plot['Month'], plot['count'], color='skyblue')
    #plt.xlabel('Last Outcome Category')
    plt.xlabel('Years')
    #plt.ylabel('Count')
    plt.title(f'City of London {gender} {od_ethnicity} ')
    plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()
```
