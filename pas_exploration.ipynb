```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
os.chdir("..")
```

```python
pas_mps_satisfaction = pd.read_excel("data/PAS_T&Cdashboard_to Q3 23-24.xlsx", sheet_name=1)
pas_borough_satisfaction = pd.read_excel("data/PAS_T&Cdashboard_to Q3 23-24.xlsx", sheet_name=2)
```

```python
pas_borough_satisfaction.columns
```

```python
grouped = pas_borough_satisfaction.groupby(["Measure"])

num_buroughs = len(pas_borough_satisfaction["Measure"].unique())
fig, axs = plt.subplots(num_buroughs, 1, figsize=(8, 8*num_buroughs), sharex=True)

# Plotting each group separately
for measure, group in grouped:
    ax = axs[[*pas_borough_satisfaction["Measure"].unique()].index(measure[0])]
    sns.boxplot(x =group["Date"], y=group["Proportion"], ax=ax)
    ax.title.set_text(measure)
    ax.set_ylim((0, 1))

# Adding labels and legend
plt.xlabel("Index")  
plt.ylabel("Proportion")
plt.legend()

# Display the plot
plt.show()
```

```python

# Assuming pas_mps_satisfaction is your DataFrame
grouped = pas_mps_satisfaction.groupby("Measure")

# Plotting each group separately
for measure, group in grouped:
    plt.plot(group["Date"], group["Proportion"], label=measure)
    plt.ylim((0, 1))

# Adding labels and legend
plt.xlabel("Date")
plt.ylabel("Proportion")
plt.legend()

# Display the plot
plt.show()
```
